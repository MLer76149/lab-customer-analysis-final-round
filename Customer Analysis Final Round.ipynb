```python
X = data_numerical_2ndorder.drop(columns=["total claim amount"])
y = data_numerical_2ndorder["total claim amount"]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MinMaxtransformer = MinMaxScaler()
MinMaxtransformer.fit(X)
x_normalized_min_max = MinMaxtransformer.transform(X)
x_normalized_min_max = pd.DataFrame(x_normalized_min_max,columns=X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_normalized_min_max, y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
lr_base = LinearRegression()
lr_base.fit(X_train, y_train)
```

```python
y_predict = lr_base.predict(X_test)
```

```python

print("R2 score: ",r2_score(y_predict, y_test))
print("MSE: ", mean_squared_error(y_predict, y_test))
print("RMSE: ", np.sqrt(mean_squared_error(y_predict, y_test)))
print("MAE: ", mean_absolute_error(y_predict, y_test))
```
